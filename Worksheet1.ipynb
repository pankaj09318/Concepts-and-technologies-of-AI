```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
A1 = np.linalg.inv(A)
print(np.round(np.matmul(A,A1)))
"""
 AA' = [[1 0]   so AA' = I proved
         [0 1]]
"""
print(np.matmul(A,B))
print(np.matmul(B,A))

"""
As AB = [[23 13]  and BA =[[36 44]  so AB != BA proved
         [51 29]]          [13 16]]
"""

print((np.matmul(A,B)).T)
A_T = A.T
B_T = B.T
print(np.matmul(B_T,A_T))

"""
here (AB)T = (BT)(AT) = [[23 51]     so (AB)T = (BT)(AT) proved
                         [13 29]]
"""


"""
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

"""

A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])
B = np.array([-1,-3,9])

print(np.dot(np.linalg.inv(A),B))
```
